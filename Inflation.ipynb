```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


yearly_since_1950 = pd.read_csv("yearly_since_1950.csv", encoding="windows-1250", sep=";", usecols=["Rok", "Wartość"])
yearly_since_1950["Wartość"] = yearly_since_1950["Wartość"].str.replace(",", ".").astype(float)
yearly_since_1950["Wartość"] -= 100
yearly_since_1950 = yearly_since_1950.rename(columns = {"Wartość": "inflation", "Rok": "year"})
year_2022 = pd.DataFrame(data={"year": [2022], "inflation": [15.6]})
yearly_since_1950 =pd.concat([yearly_since_1950, year_2022])
yearly_since_1950.head()
```

```python
# TODO: reforma monetarna

value_for_100_in_1950 = [100]
rows = yearly_since_1950.shape[0]
inflation_percents = [round(n, 2) for n in yearly_since_1950["inflation"].to_list()]
for i in range(1, rows):
    value_for_100_in_1950.append(value_for_100_in_1950[i-1]*(1+inflation_percents[i]/100))
yearly_since_1950["value_for_100_in_1950"] = value_for_100_in_1950

before_denomination = yearly_since_1950["year"] < 1995
yearly_since_1950["value_for_100_in_1950"].where(before_denomination, yearly_since_1950["value_for_100_in_1950"] / 10_000, inplace=True)

yearly_since_1950.head()
```

```python
yearly_since_1950.tail()
```

```python
sns.set(font_scale=1.2)
sns.set_theme()
sns.set_context("talk")

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(30, 20))

def palette_creator(inflation):
    if inflation <= 0:
        return "blue"
    elif inflation <= 10:
        return "white"
    return "red"

palette = [palette_creator(v) for v in yearly_since_1950["inflation"].to_list()]
sns.barplot(data=yearly_since_1950, x="year", y="inflation", ax=ax1, palette=palette)
ax1.set(yscale="symlog", yticks=[-10, 0, 10, 100], ylim=(-10, 1000), ylabel="inflation [%]")
ax1.set_title("Inflation in Poland")
ax1.tick_params(axis='x', labelrotation = 45)

def palette_creator(inflation):
    if inflation <= 100:
        return "blue"
    elif inflation <= 200:
        return "white"
    return "red"
palette = [palette_creator(v) for v in yearly_since_1950["value_for_100_in_1950"].to_list()]
sns.barplot(data=yearly_since_1950, x="year", y="value_for_100_in_1950", ax=ax2, palette=palette)
ax2.set(yscale="log", ylabel="value [PLN]")
ax2.set_title("100 PLN from 1950 value")
ax2.tick_params(axis='x', labelrotation = 45)
```
